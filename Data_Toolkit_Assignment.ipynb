#2. How does broadcasting work in NumPy?
>>> Broadcasting in NumPy refers to the ability to perform element-wise operations on arrays of different shapes, by automatically expanding the smaller array to match the shape of the larger one. This eliminates the need for explicit loops and allows for efficient computation.
For broadcasting to work, the following conditions must be met:
 a. **Matching dimensions**: The dimensions of the arrays must be either the same or one of them should be 1.
 b. **Alignment of shapes**: Starting from the rightmost dimension, the sizes must either be equal or one of them should be 1.

This allows NumPy to apply operations on arrays of different sizes without copying data, making calculations faster and more memory-efficient.

#6. What are the differences between NumPy arrays and Python lists?
>>> Here are the key differences between **NumPy arrays** and **Python lists**:

 1. **Performance**: NumPy arrays are more memory-efficient and faster for numerical operations than Python lists.
 2. **Data Type**: NumPy arrays store elements of the same data type, while Python lists can store mixed data types.
 3. **Functionality**: NumPy arrays support element-wise operations and vectorization, making mathematical operations easier, while Python lists require loops for such operations.
 4. **Size**: NumPy arrays are fixed-size, while Python lists can dynamically grow or shrink in size.

#8. What does the term “vectorized operation” mean in NumPy?
>>> A **vectorized operation** in NumPy refers to performing element-wise operations on entire arrays without using explicit loops. NumPy automatically applies the operation to all elements of the array, making the code faster and more efficient. This is achieved through optimized C and Fortran code under the hood, improving performance significantly compared to regular Python loops.

#9.  How does Matplotlib differ from Plotly?
>>> **Matplotlib** and **Plotly** are both popular data visualization libraries, but they differ in key ways:

- **Matplotlib**: It is a static plotting library, ideal for creating simple, static charts (e.g., line plots, bar charts). It offers extensive customization but is limited when it comes to interactivity.
  
- **Plotly**: It is an interactive plotting library, allowing users to create dynamic, web-based plots with zooming, hovering, and clicking features. Plotly is more suited for interactive visualizations and dashboards.

In short, Matplotlib is better for static plots, while Plotly excels in interactive visualizations.

#17.  Explain the difference between apply() and map() in Pandas?
>>> In Pandas:

- **`apply()`**: It applies a function along an axis (rows or columns) of a DataFrame or to a Series. It is more flexible and can handle functions that return a single value or multiple values.

- **`map()`**: It is used to apply a function element-wise to a Series. It is simpler and generally used for transforming or mapping individual values in a Series.

In short, `apply()` is for row/column-wise operations, while `map()` is for element-wise transformations in a Series.

#21. Why is NumPy’s array slicing faster than Python’s list slicing?
